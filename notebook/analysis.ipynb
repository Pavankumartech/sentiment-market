# Bitcoin Sentiment vs Trader Performance Analysis

This notebook analyzes the relationship between Bitcoin Market Sentiment (Fear & Greed Index) and trader performance on Hyperliquid.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Load datasets
sentiment_df = pd.read_csv('public/data/sentiment.csv')
trades_df = pd.read_csv('public/data/trades.csv')

print(f"Sentiment data shape: {sentiment_df.shape}")
print(f"Trades data shape: {trades_df.shape}")
```

## 1. Data Preprocessing
Aligning dates and cleaning the datasets.

```python
# Convert dates to datetime objects
sentiment_df['Date'] = pd.to_datetime(sentiment_df['Date'])

# Fix trade dates (handling DD-MM-YYYY format found in raw data)
def parse_trade_date(date_str):
    try:
        return pd.to_datetime(date_str, format='%d-%m-%Y')
    except:
        return pd.to_datetime(date_str)

trades_df['time'] = trades_df['time'].apply(parse_trade_date)
trades_df['Date'] = trades_df['time'].dt.normalize()

# Merge datasets
merged_df = pd.merge(trades_df, sentiment_df, on='Date', how='inner')
print(f"Merged data shape: {merged_df.shape}")
```

## 2. Analysis & Visualization
Comparing PnL and Win Rate across sentiment regimes.

```python
# Analysis: PnL by Sentiment
plt.figure(figsize=(12, 6))
sns.boxplot(x='Classification', y='closedPnL', data=merged_df)
plt.title('Profit/Loss Distribution by Market Sentiment')
plt.xticks(rotation=45)
plt.show()

# Analysis: Average Leverage by Sentiment
leverage_summary = merged_df.groupby('Classification')['leverage'].mean().sort_values()
plt.figure(figsize=(10, 5))
leverage_summary.plot(kind='bar', color='skyblue')
plt.title('Average Leverage Used per Sentiment Category')
plt.ylabel('Leverage (x)')
plt.show()
```

## 3. Conclusions
The data suggests that traders tend to use higher leverage during 'Extreme Greed' periods, which often leads to higher drawdown volatility.
